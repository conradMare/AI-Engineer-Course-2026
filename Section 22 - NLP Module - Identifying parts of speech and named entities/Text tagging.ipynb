Parts of speech tagging and named entity recognition

Two types of tagging that are going to be covered:
1 - Parts of Speech Tagging (POS)
2 - Named Entity Recognition (NER)

Parts of Speech Tagging (POS):
Take each token and assocaite them with the parts of speech (Verb, Noun etc.)

Named Entity Recognition (NER):
Instead of tagging each token, it goes through the sentence and pulls out named entities.
(NER Output: Location, Person etc. [Anything that is recognisable by a person])
