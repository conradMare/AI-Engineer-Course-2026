```python
# range(start, stop, step) -> creates a sequences of integers

# start = the first number in the list
# stop = the last value +1
# step = the distance between each two consecutive values
```

```python
range(10)
```

```python
list(range(10))
```

```python
list(range(1,10))
```

```python
list(range(2,20,3))
```
