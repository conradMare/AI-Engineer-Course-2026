```python
for n in range(10):
    print(2 ** n, end = " ")
```

```python
for x in range(20):
    if x % 2 == 0:
        print(x, end = " ")
    else:
        print("Odd", end = " ")
```

```python
x = [0, 1, 2]
for item in x:
    print(item, end = " ")
```
