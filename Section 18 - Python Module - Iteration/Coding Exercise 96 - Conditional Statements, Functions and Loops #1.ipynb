You are provided with the nums list. Define a function called count() containing a while loop to count the number of values in the nums list that are lower than 20.

```python
nums = [1,35,12,24,31,51,70,100]

def count(nums):
    total = 0
    for x in nums:
        if x < 20:
            total += 1
    return total
```

```python
count(nums)
```
