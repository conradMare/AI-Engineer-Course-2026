You have the following list of numbers. Iterate over this list, printing out each list value multiplied by 10 on the same line. Find two solutions of this problem.

```python
n = [1,2,3,4,5,6]

for num in n:
    print(num * 10, end=" ")
```

```python
n = [1,2,3,4,5,6]

for number in range(len(n)):
    print(n[number] * 10, end = " ")
```
