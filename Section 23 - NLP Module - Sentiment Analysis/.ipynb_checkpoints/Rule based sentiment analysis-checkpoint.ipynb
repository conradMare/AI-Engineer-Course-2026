```python
# Rule based sentiment analysis - also know as lexicon-based sentiment analysis.
```

```python
# Not a ML method. Instead is uses a set of predefined rules that connect words to emotions or attitudes.
```

```python
# Uses a "polarity score": a numeric value that shows whether a word is positive, negative or neutral
```

```python
# Rule-based sentiment analysis can't handle more subtle expression, such as irony or sarcasm
```

```python
# A lexicon: a collection of words paired with their predefined sentiment scores.
# Use "TextBlob" and "Vader"
# Becasue of their different lexicons, TextBlob and VADER may produce different results when anakysing the same text.
```

```python
sentence_1 = "i had a great time at the movie it was really funny"
sentence_2 = "i had a great time at the movie but the parking was terrible"
sentence_3 = "i had a great time at the movie but the parking wasn't great"
sentence_4 = "i went to see a movie"
```

TextBlob

```python
from textblob import TextBlob
```

```python
print(sentence_1)
sentiment_score_1 = TextBlob(sentence_1)
print(sentiment_score_1.sentiment.polarity)
```

```python
print(sentence_2)
sentiment_score_2 = TextBlob(sentence_2)
print(sentiment_score_2.sentiment.polarity)
```

```python
print(sentence_3)
sentiment_score_3 = TextBlob(sentence_3)
print(sentiment_score_3.sentiment.polarity)
```

```python
print(sentence_4)
sentiment_score_4 = TextBlob(sentence_4)
print(sentiment_score_4.sentiment.polarity)
```

VADER

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
vader_sentiment = SentimentIntensityAnalyzer()
```

```python
print(sentence_1)
print(vader_sentiment.polarity_scores(sentence_1))
```

```python
print(sentence_2)
print(vader_sentiment.polarity_scores(sentence_2))
```

```python
print(sentence_3)
print(vader_sentiment.polarity_scores(sentence_3))
```

```python
print(sentence_4)
print(vader_sentiment.polarity_scores(sentence_4))
```
