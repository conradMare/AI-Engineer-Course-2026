```python
import re
```

```python
# Raw strings -> use "r" in front of string
```

```python
my_folder_no_raw_string = 'C:\desktop\notes'
print(my_folder_no_raw_string)
```

```python
my_folder_raw_string = r'C:\desktop\notes'
print(my_folder_raw_string)
```

```python
# Regex search pattern -> search() checks wether a specified pattern appears within a string
```

```python
result_search = re.search('pattern', r'string to contain the pattern')
print(result_search)
```

```python
result_search_2 = re.search('pattern', r'the phrase to find is not in this string')
print(result_search_2)
```
