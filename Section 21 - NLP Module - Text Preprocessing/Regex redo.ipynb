```python
customer_review = ['sam was a great help to me in the store', 
                    'the cashier was very rude to me, I think her name was eleanor', 
                    'amazing work from sadeen!', 
                    'sarah was able to help me find the items i needed quickly', 
                    'lucy is such a great addition to the team', 
                    'great service from sara she found me what i wanted']
```

```python
sarah_reviews = []
```

```python
pattern_to_find = r'sarah?'
```

```python
for string in customer_review:
    if (re.search(pattern_to_find, string)):
        sarah_reviews.append(string)

print(sarah_reviews)
```

```python
# Note the difference if no ? was added to the pattern to find -> the ? indicates that it can be with or without a 'h'
```

```python
customer_reviews_2 = ['sam was a great help to me in the store', 
 'the cashier was very rude to me, I think her name was eleanor', 
 'amazing work from sadeen!', 
 'sarah was able to help me find the items i needed quickly', 
 'lucy is such a great addition to the team', 
 'great service from sara she found me what i wanted']
```

```python
sara_review = []
```

```python
pattern_to_find_2 = r'sara'
```

```python
for string in customer_reviews_2:
    if (re.search(pattern_to_find_2, string)):
        sara_review.append(string)

print(sara_review)

# Only one string is going to be returned
```

```python
# To find a character STARTING WITH use '^' BEFORE the character

a_reviews = []
pattern_to_find = r'^a'

for string in customer_review:
    if (re.search(pattern_to_find, string)):
        a_reviews.append(string)

print(a_reviews)
```

```python
# To find a character ENDING WITH use '$' AFTER the character

y_reviews = []
pattern_to_find = r'y$'

for string in customer_review:
    if (re.search(pattern_to_find, string)):
        y_reviews.append(string)

print(y_reviews)
```

```python
# Pipe symbol (|) works like 'or' when using regex

needwant_reviews = []
pattern_to_find = r'(need|want)ed'

for string in customer_review:
    if (re.search(pattern_to_find, string)):
        needwant_reviews.append(string)

print(needwant_reviews)
```

```python
# !!! Removing punctuation: -> VERY IMPORTANT FOR NLP (punctation interverse with word creation)

# [] -> a set of characters we want to match
# ^ -> 'not'
# \w -> word characters
# \s -> whitespaces

no_punct_reviews = []
pattern_to_find = r'[^\w\s]'

for string in customer_review:
    no_punct_string = re.sub(pattern_to_find, '', string)
    no_punct_reviews.append(no_punct_string)

print(no_punct_reviews)
```
