```python
# Lemmatization reduces a word to a meaningful base form while preserving its meaning.
# Lemmatization is more sophisticated, as it refernces a predifined dictionary to find the correct base form of a word,
# it uses context to reduce it to it's proper base form
```

```python
import nltk
from nltk.stem import PorterStemmer

ps = PorterStemmer()
```

```python
connect_tokens = ['connecting', 'connected', 'connectivity', 'connect', 'connects']
```

```python
learn_tokens = ['learns', 'learning', 'learned']
```

```python
learn_tokens_2 = ['learns', 'learning', 'learn', 'learns', 'learner', 'learners']
```

```python
likes_tokens = ['likes', 'better', 'worse']
```

```python
for t in connect_tokens:
    print(t, ' : ', ps.stem(t))
```

```python
for l in learn_tokens_2:
    print(l, ' : ', ps.stem(l))
```

```python
for lt in likes_tokens:
    print(lt, ' : ', ps.stem(lt))
```

Lemmatization

```python
nltk.download('wordnet')
# wordnet -> extensive, lexical, db of English (Basically a built-in dictionary)
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
# lemmatizer.lemmatize(t) -> takes the token 't' and looks it up in the WordNet dictionary to return its proper base form
for t in connect_tokens:
    print(t, ' : ', lemmatizer.lemmatize(t))
```

```python
# Note: the base form for lemmatization preserves the word, but this makes the dataset larger
```

```python
for l in learn_tokens_2:
    print(l, ' : ', lemmatizer.lemmatize(l))
```

```python
for lt in likes_tokens:
    print(lt, ' : ', lemmatizer.lemmatize(lt))
```
