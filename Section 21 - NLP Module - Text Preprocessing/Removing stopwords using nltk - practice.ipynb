```python
from nltk.corpus import stopwords
```

```python
en_stopwords = stopwords.words('english')
```

```python
print(en_stopwords)
```

```python
german_stopwords = stopwords.words('german')
print(german_stopwords)
```

```python
practice_sentence = 'All that glitters is not gold, and all who wander are not lost'
```

```python
sentence_no_stopwords = ' '.join([word for word in practice_sentence.split() if word not in en_stopwords])
print(sentence_no_stopwords)
```

```python
sentence_only_stopwords = ' '.join([word for word in practice_sentence.split() if word in en_stopwords])
print(sentence_only_stopwords)
```
