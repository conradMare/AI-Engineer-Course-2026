```python
new_sentence_list = ['This is a new list.',
                     'Where a NORMAL for loop is used,',
                     'INSTEAD of List Comprehesion.']
```

```python
def new_sentence_list_lower(new_sentence_list):
    lower_list = []

    for sentence in new_sentence_list:
        lower_list.append(sentence.lower())

    return lower_list
```

```python
print(new_sentence_list_lower(new_sentence_list))
```
