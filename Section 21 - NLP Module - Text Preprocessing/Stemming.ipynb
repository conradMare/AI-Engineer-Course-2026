```python
# Stemming -> reducing words to their base form
# Example: connecting or connected > connect

# Stemming reduces the number of complex words
# - Reduces complexity
# - Keeps the dataset easier to manage
```

```python
from nltk.stem import PorterStemmer
# PorterStemmer -> classic rule based algorythm that reduces English words to a simpler base
```

```python
ps = PorterStemmer()
```

```python
connect_tokens = ['connecting', 'connected', 'connectivity', 'connect', 'connects']
```

```python
for t in connect_tokens:
    print(t, ' : ', ps.stem(t))
```

```python
learn_tokens = ['learns', 'learning', 'learned']
```

```python
for l in learn_tokens:
    print(l, ' : ', ps.stem(l))
```

```python
learn_tokens_2 = ['learns', 'learning', 'learn', 'learns', 'learner', 'learners']
```

```python
for l in learn_tokens_2:
    print(l, ' : ', ps.stem(l))
```
