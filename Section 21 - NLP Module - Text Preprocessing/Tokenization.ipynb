```python
# Tokenization: breaking text into smaller unit (tokens)
```

```python
# ! Tokens have to be words, it depends on your use case -> tokens can also
# be sentences, sub-words or even characters
```

```python
import nltk
nltk.download('punkt_tab')
# 'punkt_tab' -> a table used to figure out where sentences begin and end
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
sentences = "Her cat's name is Luna. Her dogs name is Max."
```

```python
sent_tokenize(sentences)
```

```python
sentence = "Her cat's name is Luna."
word_tokenize(sentence)
```

```python
sentence_2 = "Her cat's name is Luna and her dog's name is Max"
word_tokenize(sentence_2)
```

```python
# By converting everything to lowercase, we can ensure consistency
word_tokenize(sentence_2.lower())
```
