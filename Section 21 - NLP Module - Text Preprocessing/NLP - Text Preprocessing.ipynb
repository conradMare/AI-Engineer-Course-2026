```python
sentence_list = ['Could you pass me the TV remote?',
                'It is IMPOSSIBLE to find this hotel!',
                'There once was a river in Duden...']
```

```python
lower_sentence_list = [x.lower() for x in sentence_list]
print(lower_sentence_list)
```

```python
# Example of using a normal for loop instead of list comprehension
```

```python
new_sentence_list = ['This is a list.',
                     'To LOWER case text.',
                     'Using a normal for loop instead of list comprehension.']
```

```python
def lowercase_sentence_list(new_sentence_list):
    lower_list = []
    
    for sentence in new_sentence_list:
        lower_list.append(sentence.lower())

    return lower_list

print(lowercase_sentence_list(new_sentence_list))
```
