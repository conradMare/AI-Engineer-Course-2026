```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
en_stopwords = stopwords.words('english')
```

```python
print(en_stopwords)
```

```python
sentence = 'it was to far to go to the shop and he did not want her to walk'
```

```python
sentence_no_stop_words = ' '.join([word for word in sentence.split() if word not in en_stopwords])
```

```python
print(sentence_no_stop_words)
```

```python
sentence_only_stop_words = ' '.join([word for word in sentence.split() if word in en_stopwords])
print(sentence_only_stop_words)
```

```python
en_stopwords.remove('did')
en_stopwords.remove('not')
```

```python
en_stopwords.append('go')
```

```python
sentence_no_stopwords_custom = ' '.join([word for word in sentence.split() if word not in en_stopwords])
```

```python
print(sentence_no_stopwords_custom)
```

```python
sentence_only_stop_words_custom = ' '.join([x for x in sentence.split() if x in en_stopwords])
```

```python
print(sentence_only_stop_words_custom)
```
