```python
sentence_list = ['This is a PRTACTICE sentence,',
                'To be used for lowercasing words,',
                'I am going to use THIS sentence for both:',
                'List Comprehension and,'
                'A NORMAL for loop',
                'Lookit']
```

```python
list_comprehension_lower = ' '.join([sentence.lower() for sentence in sentence_list])
```

```python
print(list_comprehension_lower)
```

```python
new_list = ['UPPERCASE',
            'lowercase']
```

```python
lower_list_comprehension = '/'.join([x.lower() for x in new_list])
```

```python
print(lower_list_comprehension)
```

```python
upper_list_comprehension = ','.join([x.upper() for x in new_list])
print(upper_list_comprehension)
```

```python
def new_lower_for(new):
    new_list = []

    for item in new:
        new_list.append(item.lower())

    return new_list
```

```python
print(new_lower_for(new_list))
```

```python
def new_upper_for(list):
    upper_list = []

    for item in list:
        upper_list.append(item.upper())

    return upper_list
```

```python
print(new_upper_for(new_list))
```
