 Function VS. Method:
 Function -> Can have many parameters
          -> Exists on its own
          -> function()

Method -> The object is one of its partents
       -> Belongs to a certain class
       -> object.method()

High level explanation of a function vs a method:
A Function is:
    -> A standalone block of code,
    -> It is not tied to any object,
    -> You call it directly by its name
    
    Example:
    def greet(name):
        print(name)

    greet("John")

A Method is:
    -> A method is a is a function that belongs to a object or a class,
    -> It is called on that object, and usually works with the objects data
    
    Example:
    text = "hello"
    text.upper()
