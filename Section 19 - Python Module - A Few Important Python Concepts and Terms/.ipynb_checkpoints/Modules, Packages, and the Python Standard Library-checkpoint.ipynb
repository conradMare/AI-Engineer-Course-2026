Module:
    Pre-Written code containing definitions of variables, functions, and classes.
    -> Can be loaded in all new programs
    ex: 
        import module

Package:
    A collection/ directory of related Python modules.
    -> The community has developed a a growing set of packages for various types of professions and specializations.
    (A Package is sometimes called a Library)
