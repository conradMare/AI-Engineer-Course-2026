```python
def raise_to_the_power_of_2(x):
    return x ** 2
```

```python
raise_to_the_power_of_2(3)
```

```python
lambda x: x ** 2
```

```python
raise_to_the_power_of_2_lambda = lambda x: x ** 2
```

```python
raise_to_the_power_of_2_lambda(3)
```

```python
(lambda x: x / 2) (11)
```

```python
(lambda x: (2 + 5 * x ** 4) ** 2 / (x  + 3) ** 3) (2)
```

```python
sum_xy = lambda x, y: x + y
sum_xy(2, 3)
```

```python
sum_xy = lambda x, y: x + y(x)
```

```python
sum_xy(2, lambda x: x * 5)
```

Lambda Functions:
- can have a one or many parameters but can contain a single expression only
- allow you to write just one line if code to include a simple functionality in a more complex expression
- can only be applied to the larger expression they have been written in
