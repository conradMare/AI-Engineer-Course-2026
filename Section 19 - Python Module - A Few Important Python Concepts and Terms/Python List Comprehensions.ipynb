List Comprehensions:
List Comprehensions are a compact and readable way to create lists in a single line of code, instead of using a traditional "for" loop.

- easy to understand
- quick to write
- support the option for setting conditionals either on the output or on the iterable OR both

```python
numbers = [1, 13, 4, 5, 63, 100]
numbers
```

```python
new_numbers = []
for n in numbers:
    new_numbers.append(n * 2)
print(new_numbers)
```

```python
# List comprehension can provide a concise way to build a new list from the given "Numbers" list
```

```python
new_numbers = [n * 2 for n in numbers]
print(new_numbers)
```

```python
for i in range(2):
    for j in range(5):
        print(i + j, end = " ")
```

```python
# Task: deliver a list containing integers equal to the values of the generated sequence raised to the power of 3 ON THE CONDITION that the base values are ODD NUMBERS
```

```python
[num ** 3 for num in range(1, 11) if num  % 2 != 0]
```

```python
# If you weren't using List Compehensions the code would like like:
```

```python
for i in range(1, 11):
    if i % 2 != 0:
        print(i ** 3, end = " ")
```

```python
[num ** 3 if num % 2 != 0 for num in range(1, 11)]
```

```python
[num ** 3 if num % 2 != 0 else 'even' for num in range(1, 11)]
```
